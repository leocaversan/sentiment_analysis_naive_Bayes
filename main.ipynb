```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text  import CountVectorizer
from sklearn.naive_bayes import MultinomialNB 

```

```python
#creating the fake base provided by gpt chat
avaliacoes_positivas = [
    "Ótimo curso!",
    "Excelente conteúdo e instrutor.",
    "Aprendizado incrível!",
    "Recomendo fortemente.",
    "Muito satisfeito com o curso.",
    "Conteúdo de alta qualidade.",
    "Instrutor experiente e bem preparado.",
    "Melhor curso que já fiz.",
    "Amei cada minuto do curso.",
    "Material didático excelente.",
    "Conteúdo atualizado e relevante.",
    "Didática clara e objetiva.",
    "Suporte e atendimento ao aluno de primeira.",
    "Um curso que realmente faz a diferença.",
    "Superou minhas expectativas.",
    "Apliquei o que aprendi imediatamente no meu trabalho.",
    "Curso indispensável para quem deseja aprender programação.",
    "Excelente custo-benefício.",
    "Conteúdo bem estruturado e organizado.",
    "Materiais complementares enriquecedores.",
    "Curso altamente recomendado.",
    "Excelente plataforma de aprendizado online.",
    "Aprendi muito e me senti motivado durante todo o curso.",
    "Clareza nas explicações e exemplos práticos.",
    "Conteúdo abrangente e bem explicado.",
    "Um dos melhores cursos da área.",
    "Fácil de acompanhar e entender.",
    "Avaliação máxima para esse curso!"
]

avaliacoes_negativas = [
    "Péssimo curso!",
    "Conteúdo fraco e mal explicado.",
    "Não recomendo.",
    "Muito insatisfeito com o curso.",
    "Material didático de baixa qualidade.",
    "Instrutor pouco preparado.",
    "Pior curso que já fiz.",
    "Detestei cada minuto do curso.",
    "Conteúdo desatualizado e irrelevante.",
    "Didática confusa e pouco objetiva.",
    "Suporte e atendimento ao aluno inexistentes.",
    "Um curso que não cumpre o prometido.",
    "Ficou aquém das minhas expectativas.",
    "Não consegui aplicar o que aprendi no meu trabalho.",
    "Curso dispensável para quem deseja aprender programação.",
    "Preço elevado para a qualidade oferecida.",
    "Conteúdo mal estruturado e desorganizado.",
    "Materiais complementares sem valor agregado.",
    "Curso não recomendado.",
    "Plataforma de aprendizado online problemática.",
    "Não aprendi nada de útil durante o curso.",
    "Explicações confusas e exemplos mal elaborados.",
    "Conteúdo superficial e mal explicado.",
    "Um dos piores cursos da área.",
    "Difícil de acompanhar e entender.",
    "Avaliação mínima para esse curso!"
]

fake_data_feedback_positive = {
    "avaliacoes": avaliacoes_positivas,
    "notas": [1 for x in range(len(avaliacoes_positivas))]
}
fake_data_feedback_negative = {
    "avaliacoes": avaliacoes_negativas,
    "notas": [0 for x in range(len(avaliacoes_negativas))]
}

df_fake_data_feedback = pd.concat([pd.DataFrame(fake_data_feedback_positive), pd.DataFrame(fake_data_feedback_negative)])
df_fake_data_feedback = df_fake_data_feedback.sample(frac = 1).reset_index(drop=True)
df_fake_data_feedback
```

```python
#separate base between training and testing
y = df_fake_data_feedback['notas']
x = df_fake_data_feedback['avaliacoes']
x, x_test, y, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

# Vectorize text reviews to numbers
vec = CountVectorizer()
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```python
#apply model
model = MultinomialNB() 
model.fit(x, y)
```

```python
#assessment model
model.score(x_test, y_test)
```
